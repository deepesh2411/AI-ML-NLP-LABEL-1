```python
##Mean squared error function
def mse(points, m, c):
    t_err = 0
    for i in range(0,len(points)):
        x=points[i,0]
        y=points[i,1]
        t_err += (y-((m*x) + c))**2

    return t_err/len(points)
```

```python
#R-SQUARE FUNCTION   
def r_square(points,m,c):
    rs, rss, tss = 0 ,0, 0

    y_mean = np.mean(points[:,1])
    
    for i in range(len(points)):
        x = points[i,0]
        y = points[i,1]
        rss += (y - (m*x + c))**2
        tss += (y - y_mean)**2

    rs = 1- (rss/tss)

    return (rs,rss,tss)
```

```python
#gradient step (slope)
def gradient(points,m_curr, c_curr,learning_rate):
    c_grad = 0
    m_grad = 0
    n = len(points)
    for i in range(0,len(points)):
        x=points[i,0]
        y=points[i,1]
        m_grad += (2/n) * (y-(m_curr * x +c_curr)) * (-x)
        c_grad += (2/n) * (y-(m_curr * x +c_curr)) * (-1)
        
    m_new = m_curr - (learning_rate * m_grad)
    c_new = c_curr - (learning_rate * c_grad)
    
    return (m_new,c_new)
```

```python
#iteration function
def gradient_iteration(points, m_start, c_start, learning_rate, iteration):
    m = m_start
    c = c_start
    
    m_list, c_list, rs_list, rss_list, tss_list = ([] for i in range(5))
    
    for i in range(iteration):
        m_prev, c_prev = m, c
        m,c = gradient(points,m, c,learning_rate)
        m_list.append(m_prev)                           #for each iteration saving the respective value to a respective list 
        c_list.append(c_prev)
        
        rs,rss,tss = r_square(points,m_prev,c_prev)
        rs_list.append(rs)
        rss_list.append(rss)
        tss_list.append(tss)
        
    return (m_prev,c_prev,m_list,c_list,rs_list,rss_list,tss_list)
    #return value are m_final, c_final, list of values of (m,c,rsquare,rss,tss) for all iterations
```
